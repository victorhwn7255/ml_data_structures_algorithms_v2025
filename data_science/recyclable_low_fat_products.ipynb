```python
import pandas as pd
```

```python
data = [
  ['0', 'Y', 'N'], 
  ['1', 'Y', 'Y'], 
  ['2', 'N', 'Y'], 
  ['3', 'Y', 'Y'], 
  ['4', 'N', 'N']
  ]
```

```python
products = pd.DataFrame(
  data, 
  columns=['product_id', 'low_fats', 'recyclable']).astype({'product_id':'int64', 'low_fats':'category', 'recyclable':'category'}
  )
```

```python
def find_products(products: pd.DataFrame):
  return products[
    (products["low_fats"] == "Y") & (products["recyclable"] == "Y")
  ][["product_id"]]
  
```

```python
print(find_products(products))
```
