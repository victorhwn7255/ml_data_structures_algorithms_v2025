```python
import pandas as pd
```

```python
data_1 = [
  [1, 'Joe'], 
  [2, 'Henry'], 
  [3, 'Sam'], 
  [4, 'Max']
  ]

customers = pd.DataFrame(data_1, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
```

```python
print(customers)
```

```python
data_2 = [
  [1, 3], 
  [2, 1]
  ]

orders = pd.DataFrame(data_2, columns=['id', 'customerId']).astype({'id':'Int64', 'customerId':'Int64'})
```

```python
print(orders)
```

```python
def find_customers(customers: pd.DataFrame, orders: pd.DataFrame):
  
  return customers[
      ~customers["id"].isin(orders["customerId"])
    ][["name"]].rename(columns={"name" : "Customers"})
```

```python
print(find_customers(customers, orders))
```
