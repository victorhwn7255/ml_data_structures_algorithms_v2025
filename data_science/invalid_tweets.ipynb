```python
import pandas as pd
```

```python
data = [
  [1, 'Let us Code'], 
  [2, 'More than fifteen chars are here!']
  ]

tweets = pd.DataFrame(data, columns=['tweet_id', 'content'])
```

```python
def invalid_tweets(tweets: pd.DataFrame):
  return tweets[
    tweets["content"].str.len() > 15
  ][["tweet_id"]]
```

```python
print(invalid_tweets(tweets))
```
