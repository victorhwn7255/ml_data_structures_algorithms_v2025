```python
import random
reader = open('../data/sherlock_holmes_1999.txt')
```

```python
successor_map = {}
window = []
```

Bigram Model -- we only predict the next word based on the previous word

To generate text, the program looks up the keys in successor_map to see what words could follow. 

```python
for line in reader:
  for word in line.split():
    clean_word = word.strip('.;,-“’”:?—‘!()_').lower()
    window.append(clean_word)
  
    if len(window) == 2:
      key = window[0]
      value = window[1]
      if key in successor_map:
        successor_map[key].append(value)
      else:
        successor_map[key] = [value]
      window.pop(0)

print(successor_map['sherlock'])
```

By randomly choosing a word from the list associated with the previous word, it can build new phrases.

```python
word = 'the'
print(word, end=' ')

for i in range(15):
  successors = successor_map[word]
  word = random.choice(successors)
  print(word, end=' ')
```

Increase context window to 2 words

```python
successor_map = {}
window = []
```

```python
for line in reader:
  for word in line.split():
    clean_word = word.strip('.;,-“’”:?—‘!()_').lower()
    window.append(clean_word)
  
    if len(window) == 3:
      key = tuple([window[0],window[1]])
      value = window[2]
      if key not in successor_map:
        successor_map[key] = [value]
      else:
        successor_map[key].append(value) 
      window.pop(0)
      
print(successor_map[("sherlock", "holmes")])
```

```python
word1 = "sherlock"
word2 = "holmes"

for i in range(15):
  print (word1, end=" ")
  successors = successor_map[(word1, word2)]
  word3 = random.choice(successors)
  word1, word2 = word2, word3
```
